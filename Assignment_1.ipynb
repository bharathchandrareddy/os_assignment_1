```python
import os

#printing the menu

print("hi What do you want to check today?")
print("Menu:")
print("1. List directory contents")
print("2. Print working directory ")
print("3. Create a new directory")
print("4. Display a message")
print("5. Concatenate text to the file and display it")
print("6. Concatenate and display content of two files")
print('7. Exit')
```

```python
d_path = 'C://Users//PC//Desktop//'
#path = "C://Users//PC//Desktop//"
#user_path = input("enter the default directory you want to choose to work with")
code = True
while code:
    
    try:
        x=int(input("enter a number from the menu 1-7 \n"))
        if x>7:
            print("Enter a number less than 7")

        elif x==1:
            try:

                directory= input("enter the directory name you want to work with \n ")
                path = os.path.join(d_path,directory)  #join the path of directory
                print(os.listdir(path))
            except:
                print("Enter a valid directory name from your computer to work with")
        elif x==2:
            print("your working directory path is '%s'" %os.getcwd())  #to print the current working directory
        elif x==3:
            try:

                directory= input("enter the directory name which you want to create \n")
                path = os.path.join(d_path,directory)
                os.mkdir(path)
                print("directory '%s' created" %directory)
            except:
                print("error in creating the directory")
        elif x==4:
            message = input('type the message you want to display')
            print(f"displayed message:{message}")  #to display the message
        elif x==5:   #opening the files, concatinating the message and to diplay the final contents of the code
            try:
                file = input('enter the file name you want to open\n')
                file_name = file+'.txt'
                file_content = input("Write what you want to add to the file\n")
                with open(file_name, 'a') as file:
                    file.write(file_content)
                    file.write('\n')
                    file.close()
                response = input('do you want to look at the contents of the file?(yes or no)')
                if response == 'yes':
                    
                    with open(file_name,'r') as file:
                        content = file.read()
                        print(content)
                        file.close()
                
                    
            except:
                print('error occured while creating the file')
                
        elif x==6:  #to play with 2 files content and add extra content to the file
            try:
                # entering the file names
                firstfile = input("Enter the name of first file ")
                secondfile = input("Enter the name of second file ")

                # opening both files in read only mode to read initial contents
                f1 = open(firstfile, 'r')
                f2 = open(secondfile, 'r')

                # printing the contents of the file before appending
                print('content of first file before appending -', f1.read())
                print('content of second file before appending -', f2.read())

                # closing the files
                f1.close()
                f2.close()

                # opening first file in append mode and second file in read mode
                f1 = open(firstfile, 'a+')
                f2 = open(secondfile, 'r')

                # appending the contents of the second file to the first file
                f1.write(f2.read())

                # relocating the cursor of the files at the beginning
                f1.seek(0)
                f2.seek(0)

                # printing the contents of the files after appendng
                print('content of first file after appending -', f1.read())
                print('content of second file after appending -', f2.read())

                # closing the files
                f1.close()
                f2.close()
                
            except:
                print('Error occured while working with the files')
            
        elif x==7:  #to break the loop
            code = False
            print("Exited")
    except:
        print("Enter a valid number")
        

```
